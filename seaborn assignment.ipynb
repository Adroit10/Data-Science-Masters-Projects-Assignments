## Question 1

Seaborn is a data visualization library based on Matplotlib that provides a high-level interface for creating informative and attractive statistical graphics. Here are five common types of plots that can be plotted using the seaborn liabrary:

1. Scatter plot:
Scatter plots are useful for visualizing the relationship between two continuous variables. They help identify patterns, trends, or correlations in the data.

2. Histogram:
Histograms are used to represent the distribution of a single variable. They help visualize the underlying frequency or probability distribution of a numeric variable.

3. Box plot (Box and whisker plot): 
Box plots are useful for visualizing the distribution of a continuous variable and identifying outliers. They provide a summary of the data's central tendency and spread.

4. Violin Plot:
Violin plots combine aspects of box plots and kernel density estimation, providing a more comprehensive view of the data distribution. They are especially useful for comparing distributions across different categories.

5. Pair plot:
Pair plots are used for exploring relationships between multiple variables in a dataset. They display scatter plots for joint relationships and histograms for individual distributions, making it easy to identify patterns and correlations.

## Question 2

```python
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset('fmri')
sns.lineplot(x=data['timepoint'],y=data['signal'])
plt.show()
```

## Question 3

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset('titanic')
plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
plt.title("pclass vs age")
sns.boxplot(x=titanic['pclass'],y=titanic['age'])
plt.subplot(1,2,2)
plt.title("pclass vs fare")
sns.boxplot(x=titanic['pclass'],y=titanic['fare'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
flights_data=sns.load_dataset('flights')
flights_data_pivot=flights_data.pivot_table(index='month',columns='year',values='passengers')
plt.figure(figsize=(15,10))
sns.heatmap(data=flights_data_pivot,annot=True)
plt.show()
```
