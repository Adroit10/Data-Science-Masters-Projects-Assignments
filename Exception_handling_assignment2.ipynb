An arithematic in python occurs when there is a error in numerical calculations. ArithematicError types include OverFlowError and ZeroDivisionError.

OverFlowError occurs whenthe result of arithematic operation is too large to be represented.

ZeroDivisionError occurs when you try to divide a number by zero.


```python
## OverFlowError
a=5
try:
    for i in range(1,1000):
        a=a**i
except OverflowError as e:
    print('overflow occured')
```

## Question 5

ImportError is an exception that is raised in Python when a module cannot be imported. This can happen for a number of reasons, such as:

1. The module does not exist.
2. The module is not installed.
3. The module is not compatible with the current version of Python.
4. There is an error in the import statement.

The 'module not found' error is a syntax error that appears when the static import statement cannot find the file at the declared path. This common syntax error is caused by letter-casing inconsistencies that are present in your filename between your repository and local machine, or both.

Exception handling is one of the most important practice in programming. A few practices that can be used in python to overcome exceptions are:

1. Use try-except-finally blocks to handle exceptions
2. Catch specific exceptions rather than using a catch all except block to avoid swallowing unexpected exceptions.
3. Raise custom exceptions with helpful descriptions.
4. Use logging of exceptions instead of print statements
