## Question 2

There are two main types of probability distributions ----

1. Discrete probability distribution 
2. Continous probability distribution
   
   
Discrete Probability Distribution:

A discrete probability distribution is a probability distribution of a categorical or discrete variable.The probability of all possible values in a discrete probability distribution add up to one. It’s certain (i.e., a probability of one) that an observation will have one of the possible values.
Various discrete probability functions include :
  Binomial disttribution, discrete uniform distribution, poisson distribution.
   
   
Continous Probability Distribution:

A continuous probability distribution is the probability distribution of a continuous variable.

A continuous variable can have any value between its lowest and highest values. Therefore, continuous probability distributions include every number in the variable’s range. A probability density function (PDF) is a mathematical function that describes a continuous probability distribution. A cumulative distribution function is another type of function that describes a continuous probability distribution.

Various distributions under Continous probability distributions include:
Normal distribution, continous uniform distribution, log normal, exponential.


## Question 3

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(1,50,100)

def normal_dist(x,mean,std):
    prob_density=(1/std*(np.sqrt(2*np.pi)))*(np.exp(-0.5*((x-mean)/std)**2))
    return prob_density

mean=np.mean(x)
std=np.std(x)
pdf=normal_dist(x,mean,std)

plt.plot(x,pdf,c='r')
plt.xlabel('Data')
plt.ylabel('Prob_density')
                                             
```
