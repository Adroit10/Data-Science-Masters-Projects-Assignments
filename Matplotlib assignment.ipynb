## Question 1

Matplotlib is a 2D plotting library for the Python programming language that produces publication-quality figures in a variety of formats and interactive environments. It is a widely used library for creating static, animated, and interactive visualizations in Python. Matplotlib provides various modules, including pyplot, which provides a MATLAB-like interface for creating a variety of plots and charts.

Five common plots that can be plotted using the matplotlib liabrary:


1. Line Plot: 
Displaying the relationship between two continuous variables over a continuous axis, often used for time series data.

2. Scatter Plot: 
Visualizing the distribution and relationship between two variables, each point represents an individual data point.

3. Bar chart: 
 Displaying the distribution of categorical data or comparing values across different categories.
 
4. Histogram:
Representing the distribution of a single variable, showing the frequency or probability of different ranges.

5. Pie chart: 
Illustrating the proportion of each category in a whole, useful for displaying parts of a whole.

## Question 2

Scatter plot is a type of visualization which is used for visualizing the distribution and relationship between two variables, each point represents an individual data point.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.title('Scatter PLot')
plt.scatter(x,y)
plt.show()
```

## Question 3

The subplot function in Matplotlib is used to create a grid of subplots within a single figure. This is particularly useful when you want to display multiple plots side by side or in a grid arrangement. The subplot function takes three integer arguments: the number of rows, the number of columns, and the index of the subplot you want to create.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0,1,2,3,4,5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,1)
plt.title('Line 1')
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.title('Line 2')
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.title('Line 3')
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.title('Line 4')
plt.plot(x,y4)
plt.tight_layout()
plt.show()
```

## Question 4

A bar plot is a graphical representation of data in which bars of rectangular shapes are used to represent the values of different categories or groups. The length of each bar corresponds to the quantity or frequency of the data it represents. Bar plots are commonly used for visualizing and comparing the values of discrete categories or groups. Bar plots are especially effective for displaying categorical data where the categories are distinct and not ordered along a continuous scale. Each bar represents a specific category, making it suitable for nominal or ordinal data.


```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.subplot(2,1,1)
plt.title("Bar plot")
plt.bar(company,profit,width=0.6)
plt.subplot(2,1,2)
plt.title("Horizontal bar plot")
plt.barh(company,profit,height=0.6)
plt.tight_layout()
plt.show()
```

## Question 5

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a summary of key statistical measures and highlights the presence of outliers. A box plot displays the following information about a dataset:

1. Median(Q2) : middle value of dataset
2. Lower and Upper Quartiles(Q1 and Q3): median of the lower and the upper half of the dataset.
3. inter Quartile Range(IQR): The range between the first and the third quartile representing the central 50% of the data.
4. Whiskers: Lines extending from the box indicating the range of data. The whiskers often extend upto 1.5 times IQR from the lower and upper whiskers.
5. Outliers: Individual data points extending the whiskers indicates outliers in our dataset.


```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(1,2,1)
plt.title("Box1")
plt.boxplot(box1)
plt.subplot(1,2,2)
plt.title("Box2")
plt.boxplot(box2)
plt.show()
```
